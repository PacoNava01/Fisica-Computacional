#### 1.- Realice una función que reciba un número y que regrese ese mismo número con cambio de signo pero con sus cifras reflejadas y conservando la posición del número decimal. Por ejemplo, si el número proporcionado es 8589.42, la función deberá regresar -2498.58.

```python
def inv():
    numero=float(input("Introduce el nùmero"))
    if numero >= 0: #Si en numero es positivo
        list_numero = str(numero) #Pasamos la entrada a una cadena de texto
        n_int = len(str(int(numero))) #La parte entera de la entrada
        n_entradas = len(str(numero))-1 #Contabilizamos el total de digitos del numero, omitiendo el punto
        s=n_entradas-n_int #Obtenemos el numero de digitos decimales despues del punto restando al total de digitos el total de la parte entera
        list_numero_inv = list_numero[::-1] #Invertimos la lista de numeros
        numero_inv = float(list_numero_inv) #Regresamos la lista a un número de tipo flotante
        numero_inv_punto = numero_inv*(10**(n_int-s)) #Mutiplicamos por potencia de 10 para reacomodar el punto decimal en su posicion 
        numero_signo = -numero_inv_punto#Cambiamos el signo delnumero invertido
        print(numero_signo)
    else:#Si el numero es negativo
        numero=numero*(-1)#Cambiamos el signo del numero y tenemos un caso analogo al caso donde el numero es positivo, por lo cual repetimoslos pasos, omitiendo el paso de volver a cmabiar el signo.
        list_numero=str(numero)
        n_int=len(str(int(numero)))
        n_entradas=len(str(numero))-1
        s=n_entradas-n_int
        list_numero_inv=list_numero[::-1]
        numero_inv=float(list_numero_inv)
        numero_inv_punto=numero_inv*(10**(n_int-s))
        print(numero_inv_punto)
```

```python
inv()
```

#### 2.- Haga las gráficas de las siguientes funciones de la función $f(x) = tan(x)$ en el intervalo $[−4\pi, 4\pi]$. La gráfica no deben tener líneas verticales.

```python
x = arange(-4*pi,4*pi,0.1) #Creamos un arreglo con los elementos entre -4pi y 4pi con incrementos de 0.1
y=tan(x) #Definimos la función de tangente con los arreglos dados en la linea anterior, esta esta completa y con lineas verticales
for i in range(len(x)):
    if y[i]-y[i-1]<0: #Los puntos que forman las rectas verticales son aquellos que su diferencia es negativa (Prueba con tan(-5)-tan(-4))
        y[i]=nan #Si su diferencia es negativa, osea, generan rectas, los eliminamos con "nan" = not a number, que no son graficados por la función plot
        
plot(x,y,label = "Tan(x)") #Graficamos nuestra nueva función tangente "modificada"
#plot(x,tan(x),"--",label = "Tan(x) completa") #sin alterar
ylim(-3,3)
grid(),legend()
show()
```

###  Genere una función y la gráfica de


$$f(x) = \left\{ \begin{array}{lr} 1 & : si x \in[n,n+1] \text{con n par}\\ 0 & : si x \in[n,n+1]\text{con n par} \end{array} \right. $$

```python
def a_b(x):
    y=zeros(len(x)) #Generamos una lista con la misma longitud de la que que un linspace(0,10,1000), donde el 10 podría ser alterado para alargar la gráfica
    for i in range(len(x)): #correremos mediante un for un "analisis" sobre los datos de las entradas de la lista mediante una revisión if-else
        entero_x = int(x[i]) #Geeneraremos una lista que ira corriendo los valores enteros de la lista linspace(0,10,1000) con la parte entero, i.e, si x=2.01--> x_int=2
        if entero_x%2 == 0: #Revisaremos que la entra de la lista x_int sea par. de serlo asignaremos un valor de 1 a la lista y[i]
            y[i]=1
        else: #En caso de que no se cumpla la primera condición, se asignara un valor de cero en esa posición 
            y[i]=0
    for i in range(len(y)): 
        #Para eliminar las lineas verticales generadas por la función plot al unir los puntos de las listas x[i] y y[i]
        # deberemos omitir los uiltimos puntos de las "plataformas" generadas, de modo que para la función y[i]
        #si la diferencia entre el punto y[i]-y[i-1] es negativo, a la entrada de dicho valor le asignaremos el valor "nan", que no es "ploteable"
        #Aprovechamos así para eliminar las lineas izquierdas y caso inverso (y[i]-y[i+1]>0) las izquierdas
        if y[i]-y[i-1]>0:   
            y[i]=nan
        elif y[i]-y[i+1]>0 :
            y[i]=nan
    #print(y)
    return y  
```

```python
x = linspace(0,10,1000)
#print(x)
y = a_b(x)
#print(y)
plot(x,y)
grid()
show()
```

### 4.-Realice la gráfica de 

$$
f(x)=\sum_{n=0}^{N}\frac{(-1)^n}{(2n+1)^2}\sin((2n+1)x) \quad \text{Para N=3,5,10,20,30}
$$

$$y[i]=f(x[i])=\sum_{n=0}^{N}\frac{(-1)^n}{(2n+1)^2}\sin((2n+1)x[i])$$

Que vendría siendo la forma en la que la maquina comprendería lo que deseamos plotear

```python
def fun():
    N = int(input("Introduce uno de los siguientes nùmeros:3,10,20,30")) #Tope de la suma
    x = linspace(-2*pi,2*pi,1000) #Generamos el dominio sobre el que corre nuestra función de 0 a 2pi en 100 pasos
    y=zeros(len(x)) #Genero una lista de elementos para la componente vertical con la misma longitud de x
    for j in range(len(x)):#Generador de elementos de la lista y
        for i in range(N+1):#Una suma de cero a N+1 para que la suma pueda dar los 4 terminos esperados sobre los elementos 
            y[j] =+ (((-1)**i)/((2*i)+1)**2)*(sin(((2*i)+1)*x[j]))
    plot(x,y)
    grid(),show()
```

```python
fun()
```

#### (a)La función de Airy Ai(x) en el intervalo [−20, 10]

La función "Airy" es una función especial, en la que la función Ai(x) y  la función relacionada Bi(x) son soluciones l.i de la ec. dif.  $$\frac{dy^2}{dx^2}-xy=0$$
Donde Ai(x) representa
$$
Ai(x)=\frac{1}{\pi\sqrt(3)}K_{1/3}(t)\\
Bi(x)=\sqrt{\frac{x}{3}}(I_{-1/3}(t)+I_{1/3}(t))\\
t=\frac{2z^{\frac{3}{2}}}{3}
$$

```python
x=linspace(-20,10,201)#Primero creamos el dominio sobre el que corre la función Airy
ai,aip,bi,bip=special.airy(x) #La función "special.airy(x)" genera 4 arreglos, dos para la función Ai(x) y otros 2 para la función Bi(x)
#Estando solo interesados en mostrar la función Ai(x), solo ploteamos dicha función
ylim(-0.5,1)
plot(x,ai,label="función Ai(x)")
plot(x,bi,label="función Bi(x)")
grid(), legend()
```

#### (b) Función de Bessel de primera especie $J_n$(x) para n = 0, 1, 2, 3.


Las funciones de Bessel son aquellas que resultan ser solución para la ec. dif, de Bessel.
$$
x^2\frac{d^2y}{dx^2}+x\frac{dx}{dy}+(x^2-\alpha^2)y=0
$$
Al igual que la función Airy, las funciones Bessel son funciones especiales de la paquetería scipy, por lo cual nuevamente debemos comenzzar por la importación de esta

```python
from scipy import * #Podemos incluir las funciones veces dentro de la paquetería si llamamos a toda esta
x = linspace(0, 10, 1000) #Generamos un dominio para nuestras bessel
#Las funciones bessel de primer orden son llamadas como jn(n,x), donde n representa el orden de la función y x la variable del dominio.
for i in range(4):
    plot(x, jn(i, x), label="$J_%d(x)$" % i)
grid()
legend();
```

#### (c) Función esférica de Bessel de primera especie $j_n(x)$ para n = 0, 1, 2, 3.

```python
from scipy import *
from scipy.special import spherical_jn
x=linspace(0,10,1000)
for i in range(4):
    plot(x,spherical_jn(i, x),label="$J_%d(x)$" % i)
grid()
legend()
show()
```
